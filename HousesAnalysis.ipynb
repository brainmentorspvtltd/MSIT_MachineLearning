```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('ggplot')
```

```python
fuel = pd.value_counts( dataset["fuel"] )
labels = fuel.index
```

```python
fuel
```

```python
fig = plt.figure(figsize=(8,5))
plt.pie( fuel, colors=["yellow", "chartreuse", "orange"], radius=1.5, explode=[0.2, 0.2, 0.2],
        shadow = True, autopct="%.2f%%", startangle=120, counterclock=False)
plt.legend(labels)
fig.savefig("pie_plot")
```

```python
writer = pd.ExcelWriter("saratoga_houses_analysis.xlsx", engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object. Note that we turn off
# the default header and skip one row to allow us to insert a user defined
# header.
dataset.to_excel(writer, sheet_name='Houses', header=True, index=False)

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Houses']

# Add a header format.
header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'border': 1})

worksheet.insert_image( 'R2', 'scatter_plot.png')
worksheet.insert_image( 'R30', 'pie_plot.png')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```
