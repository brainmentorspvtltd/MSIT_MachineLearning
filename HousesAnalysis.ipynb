```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('ggplot')
```

```python
dataset = pd.read_csv("dataset/SaratogaHouses.csv")
dataset.drop("Unnamed", axis=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset["livingArea"]
y = dataset["price"]
```

```python
plt.scatter(X,y)
plt.xlabel("Living Area")
plt.ylabel("Price")
plt.title("Price w.r.t. Living Area")
```

```python
rooms_lt_4 = dataset[ dataset["bedrooms"] < 4 ]
rooms_eq_4 = dataset[ dataset["bedrooms"] == 4 ]
rooms_gt_4 = dataset[ dataset["bedrooms"] > 4 ]
```

```python
rooms_lt_4
```

```python
rooms_eq_4
```

```python
rooms_gt_4
```

```python
fig = plt.figure(figsize=(8,5))
plt.scatter( rooms_lt_4["livingArea"], rooms_lt_4["price"], color = "red" )
plt.scatter( rooms_eq_4["livingArea"], rooms_eq_4["price"], color = "green" )
plt.scatter( rooms_gt_4["livingArea"], rooms_gt_4["price"], color = "blue" )
fig.savefig("scatter_plot")
```

```python
plt.boxplot( dataset["price"] )
```

```python
fuel = pd.value_counts( dataset["fuel"] )
labels = fuel.index
```

```python
fuel
```

```python
fig = plt.figure(figsize=(8,5))
plt.pie( fuel, colors=["yellow", "chartreuse", "orange"], radius=1.5, explode=[0.2, 0.2, 0.2],
        shadow = True, autopct="%.2f%%", startangle=120, counterclock=False)
plt.legend(labels)
fig.savefig("pie_plot")
```

```python
writer = pd.ExcelWriter("saratoga_houses_analysis.xlsx", engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object. Note that we turn off
# the default header and skip one row to allow us to insert a user defined
# header.
dataset.to_excel(writer, sheet_name='Houses', header=True, index=False)

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Houses']

# Add a header format.
header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'border': 1})

worksheet.insert_image( 'R2', 'scatter_plot.png')
worksheet.insert_image( 'R30', 'pie_plot.png')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
