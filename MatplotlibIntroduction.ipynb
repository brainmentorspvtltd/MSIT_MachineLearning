```python
deliveries.replace("Rising Pune Supergiant", "Rising Pune Supergiants", inplace=True)
```

```python
deliveries[ deliveries["batting_team"] == "Rising Pune Supergiants" ]
```

```python
caught = deliveries[ deliveries["dismissal_kind"] == "caught" ]
bowled = deliveries[ deliveries["dismissal_kind"] == "bowled" ]
```

```python
caught
```

```python
bowled
```

```python
caught_count = pd.value_counts( caught["batting_team"], sort=False )
```

```python
bowled_count = pd.value_counts( bowled["batting_team"], sort=False)
```

```python
bowled_count_sorted = []
teams = caught_count.index
for team in teams:
    bowled_count_sorted.append( bowled_count[team] )
```

```python
caught_count, bowled_count_sorted, bowled_count
```

```python
caught_count["Delhi Capitals"]
```

```python
plt.figure(figsize=(10,7))
plt.bar(teams, caught_count, color="blue")
plt.bar(teams, bowled_count_sorted, color="orange", bottom=caught_count)
plt.xticks(rotation=80)
plt.legend(("caught", "bowled"))
```
