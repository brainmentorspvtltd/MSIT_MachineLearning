```python
toss_decision = pd.value_counts( df['toss_decision'] )
```

```python
# plt.style.use('fivethirtyeight')
plt.pie(toss_decision, autopct='%.2f%%', startangle=120, 
        counterclock=False, colors=["yellow", "chartreuse"], radius=1.5, pctdistance=1.2,
       labeldistance=0.5, explode=[0.1, 0], shadow=True, rotatelabels=True )
plt.legend(["field", "bat"], loc=0)
```

```python
deliveries = pd.read_csv("dataset/ipl/deliveries.csv")
```

```python
temp_df = deliveries[ (deliveries["dismissal_kind"] != "retired hurt") & 
                     (deliveries["dismissal_kind"] != "hit wicket") & 
                     (deliveries["dismissal_kind"] != "obstructing the field") ]
```

```python
temp_df
```

```python
dismissal_kind = pd.value_counts( temp_df["dismissal_kind"] )
```

```python
dismissal_kind
```

```python
labels = dismissal_kind.index
```

```python
plt.pie(dismissal_kind, labels=labels)
```

```python
plt.style.use('ggplot')
fig = plt.figure()
fig.patch.set_facecolor("black")
plt.rcParams['text.color'] = 'red'
plt.rcParams['font.weight'] = '900'
plt.rcParams['font.family'] = 'monospace'
my_circle=plt.Circle( (0,0), 1.0, color='black')
plt.pie(dismissal_kind, autopct='%.2f%%', startangle=120, 
        counterclock=False, pctdistance=1.2, radius=1.5)
plt.legend(labels, loc=4)
p = plt.gcf().gca().add_artist(my_circle)
```
