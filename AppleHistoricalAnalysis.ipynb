```python
writer = pd.ExcelWriter("apple_historical_data_analysis.xlsx", engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object. Note that we turn off
# the default header and skip one row to allow us to insert a user defined
# header.
dataset.to_excel(writer, sheet_name='Apple', header=True, index=False)

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Apple']

# Add a header format.
header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'border': 1})

worksheet.insert_image( 'I2', 'graph.png')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
